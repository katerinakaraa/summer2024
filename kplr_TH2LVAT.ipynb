```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from astropy.io import fits 
from astropy.table import Table 
from matplotlib.colors import LogNorm 
```

```python
event_LVAT= '/Users/kate/Desktop/Summer work/kepler data and notebooks/kplr_anc-eng_TH2LVAT.fits'
hdul=fits.open(event_LVAT, mmemmap=True)
```

```python
table_hdu=hdul[1]
df=pd.DataFrame(table_hdu.data)
print(df.head())
```

```python
evt_data=Table(hdul[1].data)
evt_data
```

```python
from astropy.timeseries import LombScargle
import datetime
from datetime import datetime
```

```python
df['UTC']=pd.to_datetime(df['UTC'])
```

```python
print(df)
```

```python
full_2010 = df[df['UTC'].dt.year == 2010]
```

```python
print(full_2010)
```

```python
time = full_2010['MJD']*86400
y = full_2010['TH2LVAT']
```

```python
frequency, power = LombScargle(time, y).autopower()
```

```python
plt.plot(frequency, power)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')
plt.title('TH2LVAT All 2010 Lomb-Scargle Periodogram')
plt.show()
```
