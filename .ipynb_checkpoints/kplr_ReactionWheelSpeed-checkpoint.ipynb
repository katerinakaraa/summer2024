```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from astropy.io import fits 
from astropy.table import Table 
from matplotlib.colors import LogNorm 
from astropy.utils.data import download_file
```

```python
hdul=fits.open(event_reaction_speed, mmemmap=True)
```

```python
table_hdu=hdul[1]
df=pd.DataFrame(table_hdu.data)
print(df.head())
```

```python
evt_data=Table(hdul[1].data)
```

```python
evt_data
```

```python
from astropy.timeseries import LombScargle
import datetime
from datetime import datetime
```

```python
df['UTC']=pd.to_datetime(df['UTC'])
print(df)
```

```python
full_2010 = df[df['UTC'].dt.year == 2010]
print(full_2010)
```

```python
time = full_2010['MJD']*86400
y1 = full_2010['ADRW1SPD_']
```

```python
frequency, power = LombScargle(time, y1).autopower()
```

```python
plt.plot(frequency, power)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')
plt.title('Rxn Wheel Speeds 1 All 2010 Lomb-Scargle Periodogram')
plt.show()
```
